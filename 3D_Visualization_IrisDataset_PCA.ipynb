## 3D Visualization Iris Dataset

In this assignment students have to transform iris data into 3 dimensions
and plot a 3d chart with transformed dimensions and colour each data
point with specific class.

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn import datasets 
import numpy as np 
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt 
import plotly.express as px 
import seaborn as sns
sns.set()
```

```python
iris = datasets.load_iris()
features = iris['data']
target = iris['target']
```

```python
plt.plot(features);
```

```python
len(target)
```

```python
transformed_features = StandardScaler().fit_transform(features)
plt.plot(transformed_features);
```

```python
reduced_features = PCA(n_components=3).fit_transform(transformed_features)
plt.plot(reduced_features);
```

```python
len(reduced_features)
```

```python
fig = px.scatter_3d(reduced_features,
                    x=reduced_features[:,0], y=reduced_features[:,1], z=reduced_features[:,2],
                    color=target)
fig.show()
```
